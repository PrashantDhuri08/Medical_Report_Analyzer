```python
import pandas as pd 
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./anemia.csv')
```

```python
df.head()
```

```python
Y= df['Result']
X= df.drop(columns= ['Result'])
```

```python
xtr,xts,ytr,yts=train_test_split(X,Y, test_size=0.2,random_state=42)
```

```python
xtr
```

```python
xts.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler= StandardScaler()
```

```python
xtr_sc=xtr
xts_sc=xts
```

```python
xtr_sc.iloc[:,1:]= scaler.fit_transform(xtr.iloc[:,1:])
```

```python
xtr_sc
```

```python
xts_sc.iloc[:,1:]= scaler.fit_transform(xts.iloc[:,1:])
```

```python
from tensorflow.keras import layers
```

```python
model= keras.Sequential([
    layers.Dense(16, activation="relu", input_shape=(xtr.shape[1],)),
    layers.Dense(8, activation="relu"),
    layers.Dense(1, activation="sigmoid")
])
```

```python
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
history = model.fit(xtr_sc, ytr, validation_data=(xts_sc, yts), epochs=50, batch_size=16, verbose=1)
```
